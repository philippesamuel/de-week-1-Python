1. **FizzBuzz (classic)**

   * `fizzbuzz(n)` → `"Fizz"`, `"Buzz"`, `"FizzBuzz"`, or `str(n)`.


````python
def fizzbuzz(n) -> str:
    if (n % 3 == 0) and (n % 5 == 0):
        return "FizzBuzz"
    if (n % 3 == 0):
        return "Fizz"
    if (n % 5 == 0):
        return "Buzz"
    return str(n)
    
assert fizzbuzz(15) == "FizzBuzz"
````

3. **Truthy length**

   * `choose_label(s)` → `"empty"` if string is empty/whitespace; `"short"` (≤3), `"long"` (>3).

 

````python
def choose_label(s):
    if s.strip() == "":
      return "empty"
    if len(s) <= 3:
      return "short"
    return "long"  
````

4. **Grade boundaries**

   * `grade(score)` with 0–100 validation; raise `ValueError` otherwise. Return `A/B/C/D/F`.


````python
def grade(score) -> str:
  if not 0 <= score <= 100:
    raise ValueError(f"Grade must be in interval [0,100]")
  if score > 90:
    return "A"
  if score > 80:
    return "B"
  if score > 70:
    return "C"
  if score > 60:
    return "D"
  return "F"
````

5. **Password strength (simple)**

   * `password_strength(s)` returns `"weak"`, `"medium"`, `"strong"` based on length and presence of digit/special.

    * "weak": less than 6 characters, or only letters or only digits

    * "medium": at least 6 characters, with both letters and digits

    * "strong": at least 8 characters, with letters, digits, and special characters

 

````python
import re

def has_letter(s) -> bool:
    return bool(re.search(r"[a-zA-Z]", s))

def has_digit(s) -> bool:
    return bool(re.search(r"\d", s))

def has_special_char(s) -> bool:
    return bool(re.search(r"[^a-zA-Z0-9\s]", s))

strong_conditions = [
    lambda s: len(s)>=8,
    has_special_char,
    has_digit,
    has_letter,    
    ]

medium_conditions = [
    lambda s: len(s)>=6,
    has_digit,
    has_letter,
]

def password_strength(s):
    if all(cond(s) for cond in strong_conditions):
        return  "strong"
    if all(cond(s) for cond in medium_conditions):
        return "medium"
    return "weak"
````

8. **Safe cast**

   * `to_int(s, default=None)` returns int or default; only accept pure integer strings (e.g., `" 12 "` ok after strip, `"12.0"` not ok).

 

````python
def to_int(s, default=None):
    try:
        return int(s)
    except ValueError:
        return default
    
to_int('2.')
````

9. **Guard clauses**

   * `discount(price, code)` → if `price<=0` return 0; if `code=="NONE"` return `price`; if `code=="HALF"` return `price*0.5`; else return `price*0.9`.

 

````python
def discount(price: float, code:str) -> float:
  if price <= 0:
    return 0
  match code:
    case "NONE":
      return price
    case "HALF":
      return price * 0.5
    case _:
      return price * 0.9
  
  
discount(100, "HALF")
````

10. **Ternary map**

    * `label_numbers(nums)` → for each n: `"neg" if n<0`, `"zero" if n==0`, `"pos" if n>0`. Use a list comprehension with a ternary chain internally (but focus on if-logic correctness).

 

````python
def number2cat(n) -> str:
    return "neg" if n<0 else "zero" if n==0 else "pos" 

def label_numbers(nums) -> list[str]:    
    return list(map(number2cat, nums))

assert label_numbers([-1,0,2]) == ["neg","zero","pos"]

label_numbers([-1,0,2])
````
