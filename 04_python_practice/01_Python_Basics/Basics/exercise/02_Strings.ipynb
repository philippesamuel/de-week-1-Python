### Warm-ups

1. **Middle char(s)**

   * `middle(s)` → If odd length return the middle char, else return the two middle chars.


```python
def get_middle_idx(len: int):
    if len % 2 == 1:
        return (len - 1) // 2 # 0-indexed 
    m = len // 2
    return slice(m-1, m+1)
        

def middle(s):
    s_len = len(s)
    idx = get_middle_idx(s_len)
    return s[idx] 
    
assert middle("abc") == "b"
assert middle("abba") == "bb"
```


2. **Title-case safely**

   * `safe_title(s)` → title case but do not lowercase words fully in ALL CAPS acronyms (e.g., `"learn SQL now"` → `"Learn SQL Now"`; `"use GPU"` → `"Use GPU"`).


```python
def safe_title_word(s: str) -> str:
    if s.upper() == s:
        return s
    return s.title()

def safe_title(s: str):
    words = s.split(" ")
    titles = map(safe_title_word, words)
    return " ".join(titles)

assert safe_title("learn SQL now") == "Learn SQL Now"
assert safe_title("use GPU") == "Use GPU"
```

3. **Reverse words**

   * `reverse_words(s)` → reverse word order but keep internal characters.


```python
def reverse_words(s):
    words = s.split(" ")
    reversed_words_list = words[::-1] 
    reversed_words = " ".join(reversed_words_list)
    return reversed_words
    
assert reverse_words("one two  three") == "three  two one"
```

### Core

4. **Normalize spaces**

   * `squash_spaces(s)` → collapse multiple spaces/tabs to a single space; strip ends.


```python
def squash_spaces(s):
    _ = s.strip()
    words = _.split()
    return " ".join(words)


assert squash_spaces("  a\t b   c ") == "a b c"

print("  a\t b   c ")
print(squash_spaces('  a\t b   c '))
```


5. **CSV line → list (no quotes)**

   * `split_csv(s)` → split by commas, trim whitespace around fields; ignore empty trailing field if line ends with comma.


```python

def split_csv(s):
    splitted = s.split(",")
    stripped = map(str.strip, splitted)
    return [w for w in stripped if w]

assert split_csv(" a, b ,c,") == ["a","b","c"]
```


6. **Mask secrets**

   * `mask_email(s)` → keep first char of user and domain, mask the rest with `*`, keep TLD.

     * `"alice@example.com"` → `"a****@e******.com"`


```python
def mask_word(w: str) -> str:
    w_len = len(w)
    return w[0] + "*" * (w_len - 1)
    

def mask_email(s):
    username, domain = s.split("@")
    *base_doms, tld = domain.split(".")
    base_dom = ".".join(base_doms)
    return f"{mask_word(username)}@{mask_word(base_dom)}.{tld}"
    
assert mask_email("alice@example.com") == "a****@e******.com"
print(mask_email("alice@example.telecom.com"))
```


7. **Find all indexes**

   * `find_all(s, sub)` → list of start indices where `sub` occurs (including overlaps).


```python
def find_all(s, sub):
    pos = []
    s_len = len(s)
    for i in range(s_len):
        if s[i:].startswith(sub):
            pos.append(i)
    return pos

assert find_all("aaaa", "aa") == [0,1,2]
```


8. **Anagram check**

   * `is_anagram(a, b)` ignoring spaces, case, punctuation.[Anagram](https://en.wikipedia.org/wiki/Anagram)


```python
import re

def _ignore_non_alphanumeric(s: str) -> str:
    return re.sub(r'[^\w]', '', s).lower()

_remove_non_alnum = _ignore_non_alphanumeric

def is_anagram(a, b):
    return sorted(_ignore_non_alphanumeric(a)) == sorted(_ignore_non_alphanumeric(b))
assert is_anagram("Listen", "Silent")
```


9. **Format table row**

   * `fmt_row(values, widths)` → left-align strings to fixed widths, joined by `" | "`.

 

```python
def fmt_row(values, widths):
    fmt_cells = (v.ljust(w, " ") for v,w in zip(values, widths))
    return " | ".join(fmt_cells)

assert fmt_row(["a","bb"], [3,4]) == "a   | bb  "
```


### Challenge

10. **Slugify**

    * `slugify(title)` → lowercase, trim, replace runs of non-alnum with single `-`, remove leading/trailing `-`.


```python
def slugify(title: str) -> str:
    low_title = title.lower()
    _ = re.sub(r'[^\w]+', '-', low_title).lower()
    slug = _.strip('-')
    return slug
    
assert slugify("Hello,   World!!") == "hello-world"
```
